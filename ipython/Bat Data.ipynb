### Read SonoBat data.

```python
sonobat_file_path = os.path.join(data_dir_path, 'NFP_Pool2013_Sonobat.txt')
sonobat = pd.read_csv(sonobat_file_path, sep='\t')
```

```python
sonobat = sonobat[sonobat['Consensus'].notnull()]
```

```python
def _get_file_name_base(name):
    parts = name.split('_')
    if not parts[-1].endswith('.wav'):
        raise ValueError('File name "{}" does not end in "{}".'.format(name, '.wav'))
    parts[-1] = parts[-1][:-4].split('-')[0]
    return '_'.join(parts)
```

```python
sonobat['file_name_base'] = sonobat['Filename'].apply(_get_file_name_base)
```

### Merge KPro and SonoBat data by file name base.

```python
merged = pd.merge(kpro, sonobat, left_on='OUT FILE', right_on='file_name_base')
```

### Eliminate rows with different KPro and SonoBat classifications.

```python
def _lower(series):
    return series.apply(lambda s: s.lower())
```

```python
merged['kpro_species'] = _lower(merged['AUTO ID'])
merged['sonobat_species'] = _lower(merged['Consensus'])
merged = merged[merged['kpro_species'] == merged['sonobat_species']]
```

### Eliminate columns that do not interest us.

```python
merged = merged[['file_name_base', 'kpro_species']]
merged.rename(columns={'kpro_species': 'species'}, inplace=True)
```
