### Show some spectrograms.

```python
def show_clips(clips, clip_class_name, page_num, page_width=10, page_height=10):
    clip_class_name = 'Call.' + clip_class_name
    clips = clips[clips['clip_class'] == clip_class_name]
    grams = clips['spectra']
    jupyter_notebook_utils.show_grams(
        grams, page_num, page_width=page_width, page_height=page_height)
```

```python
show_clips(clips, 'DoubleUp', 9)
```
